```python
import numpy as np 
import visu3d as v3d 
```

```python
## tracing rays through phase optic 
from utils.phaseoptic import unif_lenslet_params, raytrace_phaseoptic, PhaseOptic, visu3d_tracer

## Camera and phase optic configuration 
cam_params = [400,400,1111]
num_lenses = 9
radius_scale = 1.5
centers, radii = unif_lenslet_params(num_lenses,cam_params,radius_scale)

element = PhaseOptic(centers, radii, mu=1.5)
out = raytrace_phaseoptic(cam_params, element)

r0, r2, r4 = out['rays_trace']
valid = out['valid_intersects']
print(f"Percentage of valid intersects: {100*valid.shape[0]/r0.shape[0]} %")
print("-------------")
```

```python
## visualizing traced rays 
idx = 10
viz = visu3d_tracer([r0[valid][::idx],r2[valid][::idx], r4[valid][::idx]])
viz_all = visu3d_tracer([r0[::idx],r2[::idx], r4[::idx]])
```
